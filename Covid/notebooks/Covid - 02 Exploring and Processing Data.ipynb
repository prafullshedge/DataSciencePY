```python
import os
import pandas as pd
import numpy as np
```

```python
#set the path of the raw data
raw_folder_path = 'C:/Users/shedgep/DataScience/Covid/raw'
AgeGroupDetails_file_path = os.path.join(raw_folder_path,'AgeGroupDetails.csv')
HospitalBedsIndia_file_path = os.path.join(raw_folder_path,'HospitalBedsIndia.csv')
ICMRTestingLabs_file_path = os.path.join(raw_folder_path,'ICMRTestingLabs.csv')
IndividualDetails_file_path = os.path.join(raw_folder_path,'IndividualDetails.csv')
StatewiseTestingDetails_file_path = os.path.join(raw_folder_path,'StatewiseTestingDetails.csv')
covid_19_india_file_path = os.path.join(raw_folder_path,'covid_19_india.csv')
population_india_census2011_file_path = os.path.join(raw_folder_path,'population_india_census2011.csv')
```

```python
#read the data with all default parameters
AgeGroupDetails_df = pd.read_csv(AgeGroupDetails_file_path, index_col='Sno')
HospitalBedsIndia_df = pd.read_csv(HospitalBedsIndia_file_path, index_col='Sno')
ICMRTestingLabs_df = pd.read_csv(ICMRTestingLabs_file_path, index_col='lab')
IndividualDetails_df = pd.read_csv(IndividualDetails_file_path, index_col='id')
StatewiseTestingDetails_df = pd.read_csv(StatewiseTestingDetails_file_path)
covid_19_india_df = pd.read_csv(covid_19_india_file_path, index_col='Sno')
population_india_census2011_df = pd.read_csv(population_india_census2011_file_path, index_col='Sno')
```

```python
#Renaming column
covid_19_india_df = covid_19_india_df.rename(columns = {'State/UnionTerritory':'State'})
population_india_census2011_df = population_india_census2011_df.rename(columns = {'State / Union Territory':'State'})
HospitalBedsIndia_df = HospitalBedsIndia_df.rename(columns = {'State/UT':'State'})
```

```python
#Changing date folrmat, sometime it was swaping day with month
#covid_19_india_df['Date'] = pd.to_datetime(covid_19_india_df['Date']).dt.strftime("%d/%m/%Y")
#StatewiseTestingDetails_df['Date'] = pd.to_datetime(StatewiseTestingDetails_df['Date']).dt.strftime("%d/%m/%Y")
#Its working fine as it format date with yyy-mm-dd
covid_19_india_df['Date'] = pd.to_datetime(covid_19_india_df.Date, dayfirst=True)
StatewiseTestingDetails_df['Date'] = pd.to_datetime(StatewiseTestingDetails_df.Date, dayfirst=True)
```

```python
#Join the Dataframes on defined columns
main_df = covid_19_india_df.merge(StatewiseTestingDetails_df,how='left',left_on=['Date','State'],right_on=['Date','State'])
main_df = main_df.merge(population_india_census2011_df,how='left',left_on=['State'],right_on=['State'])
main_df = main_df.merge(HospitalBedsIndia_df,how='left',left_on=['State'],right_on=['State'])
#Sorting the values on State and Date
main_df = main_df.sort_values(["State","Date"], ascending=(True,True))
main_df.info()
```
