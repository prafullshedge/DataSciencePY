```python
import numpy as np
import pandas as pd
import pandas_bokeh
```

```python
df = pd.DataFrame({
    'Year': np.arange(2010, 2020),
    'Category-A': np.random.uniform(9000,15000,10),
    'Category-B': np.random.uniform(9000,15000,10),
    'Category-C': np.random.uniform(9000,15000,10)
})
```

```python
df.head(100)
```

```python
pandas_bokeh.output_notebook()
```

```python
df.plot_bokeh(
kind='bar',
x='Year',
y=['Category-A','Category-B','Category-C'],
xlabel='Category',
ylabel='Annual Sales',
title='Annual Sales by Category'
)
```

```python
df.plot_bokeh(
kind='line',
x='Year',
y=['Category-A','Category-B','Category-C'],
figsize=(900, 500),
ylim=(5000, 20000),
zooming=False,
panning=False
)
```

```python
pandas_bokeh.output_file('chart.html')
```
